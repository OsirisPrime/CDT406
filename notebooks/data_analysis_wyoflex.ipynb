## Plotting Average of Restored Data with Frequencies Removed

```python
import numpy as np
from scipy.signal import butter, filtfilt
import matplotlib.pyplot as plt

# Set your sampling frequency (replace with your actual value)
fs = 1000  # Hz

# Design Butterworth bandpass filter
lowcut = 20
highcut = 150
order = 4
b, a = butter(order, [lowcut / (0.5 * fs), highcut / (0.5 * fs)], btype='band')

# Apply filter to each row in restored_df
filtered_signals = restored_df.apply(lambda row: filtfilt(b, a, row.values), axis=1)
filtered_signals = np.vstack(filtered_signals.values)

# Compute and plot the average filtered signal
mean_filtered_signal = filtered_signals.mean(axis=0)

plt.figure(figsize=(12, 4))
plt.plot(mean_filtered_signal)
plt.title('Average EMG Signal (20–150 Hz Bandpassed)')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.show()
```

# Outlier Investigation

```python
import numpy as np
import matplotlib.pyplot as plt

# Compute mean and std for each signal (row)
signal_means = restored_df.mean(axis=1)
signal_stds = restored_df.std(axis=1)

# Define outlier threshold (e.g., 3 standard deviations from the mean)
mean_threshold = signal_means.mean() + 1 * signal_means.std()
std_threshold = signal_stds.mean() + 1 * signal_stds.std()

# Find outlier indices
outlier_indices = restored_df.index[(signal_means > mean_threshold) | (signal_stds > std_threshold)]

print(f"Outlier indices: {list(outlier_indices)}")

# Plot a few outlier signals
for idx in outlier_indices[:5]:
    plt.figure(figsize=(12, 4))
    plt.plot(restored_df.iloc[idx].values)
    plt.title(f'Outlier Signal at Index {idx}')
    plt.xlabel('Time (samples)')
    plt.ylabel('Amplitude')
    plt.show()
```

```python
import matplotlib.pyplot as plt

# Compute the mean signal before outlier removal
mean_before = restored_df.mean(axis=0).values

# Remove outlier rows
filtered_df = restored_df.drop(index=outlier_indices)

# Compute the mean signal after outlier removal
mean_after = filtered_df.mean(axis=0).values

# Plot both averages
plt.figure(figsize=(12, 4))
plt.plot(mean_before, label='Before Outlier Removal', color='blue')
plt.plot(mean_after, label='After Outlier Removal', color='orange')
plt.title('Average EMG Signal Before and After Outlier Removal')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt

# Bandpass filter parameters
fs = 1000  # Hz
lowcut = 20
highcut = 150
order = 4
b, a = butter(order, [lowcut / (0.5 * fs), highcut / (0.5 * fs)], btype='band')

# Filter all signals (before outlier removal) and take absolute value
filtered_signals_before = restored_df.apply(lambda row: np.abs(filtfilt(b, a, row.values)), axis=1)
filtered_signals_before = np.vstack(filtered_signals_before.values)
mean_abs_filtered_before = filtered_signals_before.mean(axis=0)

# Remove outliers and filter again, then take absolute value
filtered_df = restored_df.drop(index=outlier_indices)
filtered_signals_after = filtered_df.apply(lambda row: np.abs(filtfilt(b, a, row.values)), axis=1)
filtered_signals_after = np.vstack(filtered_signals_after.values)
mean_abs_filtered_after = filtered_signals_after.mean(axis=0)

# Plot both averages (absolute values)
plt.figure(figsize=(12, 4))
plt.plot(mean_abs_filtered_before, label='Before Outlier Removal (Abs Filtered)', color='blue')
plt.plot(mean_abs_filtered_after, label='After Outlier Removal (Abs Filtered)', color='orange')
plt.title('Average Absolute Filtered EMG Signal Before and After Outlier Removal')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude (Absolute)')
plt.legend()
plt.show()
```

## Histogram of Restored Signals

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000  # Hz
n = restored_df.shape[1]
freqs = np.fft.rfftfreq(n, d=1/fs)

# Compute FFT magnitude for each signal
fft_magnitudes = restored_df.apply(lambda row: np.abs(np.fft.rfft(row.values)), axis=1)
fft_magnitudes = np.vstack(fft_magnitudes.values)

# Average magnitude spectrum
mean_magnitude = fft_magnitudes.mean(axis=0)

mean_magnitude = mean_magnitude[1:]
freqs = freqs[1:]

# Plot as a line plot
plt.figure(figsize=(12, 6))
plt.plot(freqs, mean_magnitude, color='blue')
plt.title('Average Frequency Magnitude Spectrum (Restored Data)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Average Magnitude')
plt.xlim(0, fs/2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000  # Hz
n = restored_df.shape[1]
freqs = np.fft.rfftfreq(n, d=1/fs)

# Compute FFT magnitude for each signal
fft_magnitudes = restored_df.apply(lambda row: np.abs(np.fft.rfft(row.values)), axis=1)
fft_magnitudes = np.vstack(fft_magnitudes.values)

threshold = 1.5

# Count how many times each frequency bin exceeds the threshold
freq_counts = (fft_magnitudes > threshold).sum(axis=0)

plt.figure(figsize=(12, 6))
plt.bar(freqs, freq_counts, width=freqs[1]-freqs[0], color='skyblue', edgecolor='k')
plt.title('Number of Signals with Significant Magnitude at Each Frequency')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Count')
plt.show()
```
